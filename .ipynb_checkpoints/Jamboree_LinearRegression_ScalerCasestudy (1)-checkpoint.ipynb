```python
sns.set_theme(style = 'dark')
def feature_countplot(ax, data, feature, title='Count Plot', xlabel=None, ylabel='Count', palette='viridis'):
    sns.countplot(data=data, x=feature, palette=palette, ax=ax)
    # Adding title and labels
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel(xlabel if xlabel else feature, fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)
    # Customizing ticks
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)
    # Adding value annotations on top of the bars
    for p in ax.patches:
        ax.annotate(f'{p.get_height()}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom', fontsize=10)
```

```python
def plot_distribution_and_box(ax, data, feature):
    # Distribution Plot (KDE)
    sns.kdeplot(data[feature], ax=ax[0], color='orange', fill=True, alpha=0.5, linewidth=1)
    ax[0].set_title(f'Distribution of {feature}', fontsize=8)
    ax[0].set_xlabel(feature, fontsize=6)
    ax[0].set_ylabel('Density', fontsize=6)

    # Box Plot
    sns.boxplot(x=data[feature], ax=ax[1], color='lightgray', palette='rocket_r')
    ax[1].set_title(f'Box Plot of {feature}', fontsize=8)
    ax[1].set_xlabel(feature, fontsize=6)
```

```python
X_train = pd.DataFrame(X_train, columns=X_train_columns)
X_test = pd.DataFrame(X_test, columns=X_train_columns)
X_train.head()
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
print(f"Coefficients: {lr_model.coef_}")
print(f"Intercept: {lr_model.intercept_}")
```

```python
def model_scores(X, y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    adjusted_r2 = 1 - (1 - r2) * (len(y_true) - 1) / (len(y_true) - X.shape[1] - 1)

    print(f"Root Mean Squared Error: {rmse:.2f}")
    print(f"R-squared: {r2:.2f}")
    print(f"Adjusted R-squared: {adjusted_r2:.2f}")
```

```python
def model_evaluation(model, X_train, y_train, X_test, y_test):
    print("Training Set Evaluation:")
    y_train_pred = model.predict(X_train)
    model_scores(X_train, y_train, y_train_pred)
    print("\nTest Set Evaluation:")
    y_test_pred = model.predict(X_test)
    model_scores(X_test, y_test, y_test_pred)

    return y_train_pred, y_test_pred
```

```python
y_train_pred, y_test_pred = model_evaluation(lr_model, X_train, y_train, X_test, y_test)
```

Here the accuracy for training and testing data is close to 81.

```python
sns.set_theme(style='dark')
def plot_actual_vs_predicted_and_residuals(y_test, y_pred, model_name='Lasso'):
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    sns.regplot(x=y_test, y=y_pred, ax=axs[0], scatter_kws={'alpha': 0.6, 's': 50, 'color': 'blue'},
                line_kws={'color': 'red'}, lowess=True)

    axs[0].set_title(f"Actual vs. Predicted ({model_name} Linear Regression)", fontsize=12, fontweight='bold')
    axs[0].set_xlabel("Actual Chance of Admit", fontsize=12)
    axs[0].set_ylabel("Predicted Chance of Admit", fontsize=12)

    # Calculate residuals
    residuals = y_test - y_pred

    axs[1].scatter(y_test, residuals, alpha=0.6, color='blue', edgecolor='k', s=50)
    axs[1].axhline(0, color='red', linestyle='--')

    axs[1].set_title(f'Actual vs. Residuals ({model_name})', fontsize=12, fontweight='bold')
    axs[1].set_xlabel('Actual Values', fontsize=12)
    axs[1].set_ylabel('Residuals', fontsize=12)

    axs[1].grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_actual_vs_predicted_and_residuals(y_test, y_test_pred, model_name='Base Model')
```

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

ols_model = sm.OLS(np.array(y_train), X_train_sm).fit()
print(ols_model.summary())
```

Here the probability of F-statistic is very low which indicates that the model is highly significant.

**Total score** with p-value of 0 is the most significant predictor. The coefficient of 0.1028 indicates that for every one-unit increase in Total Score, the dependent variable y is expected to increase by 0.1028 units, holding all other variables constant.

**LOR** with p-value of 0 is the most significant predictor. The coefficient of 0.0186 indicates that for every one-unit increase in Total Score, the dependent variable y is expected to increase by 0.0186 units, holding all other variables constant.

**Research** with p-value of 0.001 is the most significant predictor. The coefficient of 0.0126 indicates that for every one-unit increase in Total Score, the dependent variable y is expected to increase by 0.0126 units, holding all other variables constant.

**University Rating (p-value = 0.507) and SOP (p-value = 0.594)**, are not statistically significant at a 0.05 level. This means that based on this model, there is no reliable evidence to conclude that these variables have a meaningful linear relationship with the dependent variable y.

#Model Diagnostics

**The Durbin-Watson statistic is 2.090**, which is very close to 2. This suggests that there is no significant autocorrelation among the residuals, which is a good sign.

**The Omnibus and Jarque-Bera tests both have p-values of 0.000**. This indicates that the residuals of the model are not normally distributed, a violation of a key assumption of OLS regression. The negative Skew (-1.098) suggests a left-skewed distribution of residuals, while the high Kurtosis (5.277) suggests a more heavy-tailed distribution (more outliers) than a normal distribution.

**The Cond. No. (Condition Number) is 3.56**, a low value which suggests that there is no significant multicollinearity among the predictor variables.

# Assumptions of Linear Regression

1. Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data
```

```python
vif_scores = calculate_vif(X_train)
print(vif_scores)
```

**There is no significant multicollinearity among the independent variables**. Here the VIF values are ranging from 1.4 to 2.9. It means each independent variable is contributing unique information to the model, allowing for a more stable and reliable interpretation of the regression coefficients.

2. Linearity of Variables

```python
residuals = y_test - y_test_pred

sns.set_theme(style='whitegrid')

plt.figure(figsize=(8, 5))
plt.scatter(y_test_pred, residuals, alpha=0.8, color='blue', edgecolor='k', s=50)

plt.axhline(0, color='red', linestyle='--')

plt.title('Residuals vs. Predicted Values', fontsize=10, fontweight='bold')
plt.xlabel('Predicted Values', fontsize=8)
plt.ylabel('Residuals', fontsize=8)

plt.grid(False)
plt.show()
```

An ideal residuals plot would show the points randomly scattered around the horizontal zero line, with a constant spread. The pattern observed here suggests that the model's predictive accuracy is less consistent across the range of the predicted values.

**This unequal variance of residuals is a strong indication of heteroscedasticity.**

3. Heteroscedasticity

Let's see if heteroscedasticity is there or not.

```python
from statsmodels.stats.diagnostic import het_goldfeldquandt

gq_test = het_goldfeldquandt(ols_model.resid, ols_model.model.exog)
gq_test_statistic = gq_test[0]
gq_p_value = gq_test[1]

print("Goldfeld-Quandt Test Statistic:", gq_test_statistic)
print("p-value:", gq_p_value)
```

```python
if gq_p_value > 0.05:
    print("Fail to reject the null hypothesis: No evidence of heteroscedasticity.")
else:
    print("Reject the null hypothesis: Heteroscedasticity may be present.")
```

The null hypothesis of the Goldfeld-Quandt test is that the variance of the residuals is equal.

The test statistic of 1.008 is very close to 1.0, indicating that the variance of the residuals in the two sub-samples created by the test is nearly identical.

4. Normality of Errors

```python
sns.set_theme(style='dark')
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Histogram of Residuals
sns.histplot(residuals, bins=30, kde=True, color='blue', ax=axs[0])
axs[0].set_title('Histogram of Residuals', fontsize=10)
axs[0].set_xlabel('Residuals', fontsize=8)
axs[0].set_ylabel('Frequency', fontsize=8)
axs[0].grid(False)

# Q-Q Plot
stats.probplot(residuals, dist="norm", plot=axs[1])
axs[1].get_lines()[1].set_color('orange')
axs[1].get_lines()[0].set_markerfacecolor('red')
axs[1].set_title('Q-Q Plot of Residuals', fontsize=10)
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

Here the histogram of residuals are not perfectly bell-shaped. The plot appears to be slightly left-skewed or negatively skewed.

Q-Q plot is the clear deviation from the straight line. This S-shaped curve in the Q-Q plot is a strong indicator that the residuals are not normally distributed.

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train, y_train)
```

```python
y_lasso_train_pred, y_lasso_test_pred = model_evaluation(lasso_model, X_train, y_train, X_test, y_test)
```

```python
plot_actual_vs_predicted_and_residuals(y_test, y_lasso_test_pred, model_name='Lasso')
```

Here, both the low R-squared values on both the training set (0.28) and the test set (0.29) are same. An R-squared of 0.29 means the 29% of the variance in the "Chance of Admit" variable, which is a low value. The Root Mean Squared Error (RMSE) is 0.12 for both sets which may be considered high depending on the scale of the target variable.

Actual vs Predicted plot shows that it is a poor fit.

While there's a general positive trend, the data points are widely scattered around the red regression line, showing that the model's predictions are often far from the actual values.

Actual vs Residuals show that it is non-linear.The residuals are not randomly scattered around the zero line; they form a U-shaped curve. So, the Lasso model is an inappropriate choice.

**The model suffers from high bias and likely low variance.**

**High Bias:** The poor R-squared and the clear U-shaped pattern in the residuals plot both point to high bias. The model is too simple and is systematically underfitting the data because it assumes a linear relationship where a non-linear one likely exists.

**Low Variance:** The evaluation metrics on the training set and test set are very similar (R-squared of 0.28 vs. 0.29, and RMSE of 0.12 vs. 0.12). This indicates the model performs consistently on unseen data and is not overfitting.

**So, the model is underfitting the data.**
